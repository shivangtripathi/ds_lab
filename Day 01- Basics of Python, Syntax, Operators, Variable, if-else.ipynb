# 2: Different types of Operators

## 2.1 Arithmetic Operators (+, -, \*, /, //, %, **)

### 2.1.1: Use of + - * / %

```python
#------------------------------------------------------------------------
# 2.1.1: Use of + - * / %
#------------------------------------------------------------------------
a = 220
b = 16

c = a + b       # Use of +
print (a, "+", b, "-->", c)

c = a - b       # Use of -
print (a, "-", b, "-->", c)

c = a * b       # Use of *
print (a, "*", b, "-->", c)

c = a / b       # Use of /
print (a, "/", b, "-->", c)

c = a % b       # Use of %
print (a, "%", b, "-->", c)
```

### 2.1.2: Use of / and  //

```python
#------------------------------------------------------------------------
# 2.1.2: Use of / and  //
#------------------------------------------------------------------------
a = 220
b = 16

c = a / b       # Use of /
print (a, " /", b, "-->", c)

c = a // b      # Use of //
print (a, "//", b, "-->", c)
```

### 2.1.3: Use of ** (power)

```python
#------------------------------------------------------------------------
# 2.1.3: Use of ** (power)
#------------------------------------------------------------------------
a = 9
b = 2

c = a ** b       # Use of **
print (a, "**", b, "-->", c)
```

## 2.2: Bitwise Operators (&, |, ~, ^, >>, <<)

### 2.2.1: Use of & (and)

```python
#------------------------------------------------------------------------
# 2.2.1: Use of & (and) --> Sets each bit to 1 if both bits are 1.
#------------------------------------------------------------------------
#Example 1:
a = 0
b = 1
d = 1

c = a & b       # Use of & 
print (a, "&", b, "-->", c)

c = d & b       # Use of &
print (d, "&", b, "-->", c)
```

```python
#Example 2:
a = 20
b = 10
d = 10

c = a & b       # Use of & 
print (a, "&", b, "-->", c)

c = d & b       # Use of &
print (d, "&", b, "-->", c)
```

### 2.2.2: Use of | (or)

```python
#------------------------------------------------------------------------
# 2.2.2: Use of | (or) --> Sets each bit to 1 if one of two bits is 1.
#------------------------------------------------------------------------
#Example 1:
a = 0
b = 1
d = 1

c = a | b       # Use of |
print (a, "|", b, "-->", c)

c = d | b       # Use of |
print (d, "|", b, "-->", c)
```

```python
#Example 2:
a = 20
b = 10
d = 10

c = a | b       # Use of |
print (a, "|", b, "-->", c)

c = d | b       # Use of |
print (d, "|", b, "-->", c)
```

### 2.2.3: Use of ~ (not)

```python
#------------------------------------------------------------------------
# 2.2.3: Use of ~ (not) --> Inverts all the bits.
#------------------------------------------------------------------------
#Example 1:
a = 0
b = 1

c = ~a        # Use of ~
print ("~", a, "-->", c)

c = ~b       # Use of ~
print ("~", b, "-->", c)
```

```python
#Example 2:
a = 20
b = 10

c = ~a        # Use of ~
print ("~", a, "-->", c)

c = ~b       # Use of ~
print ("~", b, "-->", c)
```

### 2.2.4: Use of ^ (xor)

```python
#------------------------------------------------------------------------
# 2.2.4: Use of ^ (xor) --> Sets each bit to 1 if only one of two bits is 1.
#------------------------------------------------------------------------
#Example 1:
a = 0
b = 1
d = 1

c = a ^ b       # Use of ^
print (a, "^", b, "-->", c)

c = d ^ b       # Use of ^
print (d, "^", b, "-->", c)
```

```python
#Example 2:
a = 20
b = 10
d = 10

c = a ^ b       # Use of ^
print (a, "^", b, "-->", c)

c = d ^ b       # Use of ^
print (d, "^", b, "-->", c)
```

### 2.2.5: Use  >> (shift right)

```python
#------------------------------------------------------------------------
# 2.2.5: Use  >> (shift right) --> Shift right by pushing copies of the leftmost 
# bit in from the left, and let the rightmost bits fall off.
#------------------------------------------------------------------------
#Example 1:
a = 1
b = 1
d = 2

c = a >> b       # Use of >>
print (a, ">>", b, "-->", c)

c = a >> d       # Use of >>
print (a, ">>", d, "-->", c)
```

```python
#Example 2:
a = 20
b = 1
d = 2

c = a >> b       # Use of >>
print (a, ">>", b, "-->", c)

c = a >> d       # Use of >>
print (a, ">>", d, "-->", c)
```

### 2.2.6: Use  << (shift left)

```python
#------------------------------------------------------------------------
# 2.2.6: Use  << (shift left) --> Zero fill left shift
# Shift left by pushing zeros in from the right and let the leftmost bits fall off.
#------------------------------------------------------------------------
#Example 1:
a = 1
b = 1
d = 2

c = a << b       # Use of <<
print (a, "<<", b, "-->", c)

c = a << d       # Use of <<
print (a, "<<", d, "-->", c)
```

```python
#Example 2:
a = 20
b = 1
d = 2

c = a << b       # Use of <<
print (a, "<<", b, "-->", c)

c = a << d       # Use of <<
print (a, "<<", d, "-->", c)
```

### 4.5: Identity Operators (is, is not)

```python
#------------------------------------------------------------------------
# 4.5: Identity Operators (is, is not)
#------------------------------------------------------------------------
a = 10
b = 20
print ("a is a     -->", a is a)
print ("a is not a -->", a is not a)
print("")
print ("a is b     -->", a is b)
print ("a is not b -->", a is not b)
```
